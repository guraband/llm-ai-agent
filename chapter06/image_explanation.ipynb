```python
import base64

def encode_image(image_path: str):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

image_path = "./images/oecd_rnd_2021_medium.png"
encoded_image = encode_image(image_path)

print(encoded_image)
```

```python
oecd_rnd_2021_base64 = encode_image("./images/oecd_rnd_2021_large.png")
oecd_rnd_2022_base64 = encode_image("./images/oecd_rnd_2022.png")

messages = [
    {
        "role": "user",
        "content": [
            {"type": "text", "text": """첫번째 이미지는 2021년 데이터이고 두번째 이미지는 2022년 데이터야. 
             어떤 데이터이고, 한국 데이터에 어떤 변화가 있었는지 설명해줘."""},
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/png;base64,{oecd_rnd_2021_base64}"
                }
            },
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/png;base64,{oecd_rnd_2022_base64}"
                }
            }
        ]
    }
]

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
)

response.choices[0].message.content
```

위의 두 이미지는 2021년과 2022년의 각국의 연구개발비와 GDP 대비 연구개발비 비율을 비교한 데이터입니다. 

### 데이터 변화 요약:

1. **연구개발비 (백만 US 달러)**:
   - **2021년**: 한국의 연구개발비는 약 89,282백만 달러.
   - **2022년**: 한국의 연구개발비는 약 87,225백만 달러로 감소.

2. **GDP 대비 연구개발비 비율 (%)**:
   - **2021년**: 한국의 비율은 4.93%.
   - **2022년**: 한국의 비율은 2.91%로 감소.

### 한국 데이터 변화 분석:

- **연구개발비 감소**: 한국의 연구개발비가 2021년에서 2022년으로 넘어가는 동안 감소하였습니다. 이는 연구개발 투자에 대한 정부 또는 기업의 예산 조정이 있었음을 시사할 수 있습니다.
  
- **비율 감소**: GDP 대비 연구개발비 비율 또한 크게 감소했습니다. 이는 전체 경제 성장률이 연구개발비 증가보다 빠르게 증가했음을 나타낼 가능성이 있습니다.
